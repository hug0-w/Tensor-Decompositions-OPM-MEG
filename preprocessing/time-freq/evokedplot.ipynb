```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
```

```python
mat_path = r'Outcome\avg_outcome_plot_data.mat'
m = loadmat(mat_path)
time_outcome = np.squeeze(m["t"])         
pl_outcome = m["pl"]                
pl_outcome = np.array(pl_outcome)
```

```python
mat_path = r'Choice\avg_choice_plot_data.mat'
m = loadmat(mat_path)
time_choice = np.squeeze(m["t"])         
pl_choice = m["pl"]                
pl_choice = np.array(pl_choice)
```

```python
plt.figure(figsize=(16,8))

plt.plot(time_choice,pl_choice);
plt.grid(ls='--')
plt.ylabel("B (fT)")
plt.xlabel('Time (s)')
plt.vlines(0,-800,800,color='r')
plt.title('Evoked Response - Choice');
plt.ylim(-800,800)
plt.xlim(-0.5,1.5)
```

```python
plt.figure(figsize=(16,8))

plt.plot(time_outcome,pl_outcome);
plt.grid(ls='--')
plt.ylabel("B (fT)")
plt.xlabel('Time (s)')
plt.vlines(0,-800,800,color='r')
plt.title('Evoked Response - Outcome');
plt.ylim(-800,800)
plt.xlim(-0.5,1)
```

```python
fig, ax = plt.subplots(2,1, figsize=(16,8),sharex=True)

times = [time_choice, time_outcome]
pls = [pl_choice, pl_outcome]
titles = ['Choice', 'Outcome']

for idx,i in enumerate(times):
    
    ax[idx].plot(times[idx],pls[idx])
    ax[idx].set_title(titles[idx])
    ax[idx].set_ylabel('B (fT)')
    ax[idx].set_xlabel('Time (s)')
    ax[idx].grid(ls='--')
    ax[idx].vlines(0,-800,800,color='r')
plt.tight_layout()
```
